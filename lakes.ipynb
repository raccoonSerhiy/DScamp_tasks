```python
import numpy as np 
import matplotlib.pyplot as plt

size=20
np.random.seed(2023)
a = np.random.randint(1,100,size)

plt.plot (a)

l,r=0,19
depth=0

lake= True
prev=0
for i in range(size-1):
    if lake and a[i]>=a[i+1]: continue
    lake=False
    if a[i]<=a[i+1]:continue
    lake = True

    cur_depth=min(a[prev],a[i])-np.min(a[prev:i])
    if cur_depth>depth:
        depth=cur_depth
        l,r=prev,i
    prev=i    
#перевірка правого краю    
cur_depth=min(a[prev],a[size-1])-np.min(a[prev:size-1])
if cur_depth>depth:
        depth=cur_depth
        l,r=prev,size-1
    
   
print(f"лівий берег - {l}, правий берег - {r}, глибина - {depth}")
    
plt.plot(range(l,r+1),a[l:r+1], c='red')
```
